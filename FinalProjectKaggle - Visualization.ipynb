# Imports

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt      
import seaborn as sns
```

```python
data_of_database_all_pages = pd.read_csv('All_Data_Stored_From_Kaggle_after_cleaning.csv')
data_of_database_all_pages.drop('Unnamed: 0', inplace=True, axis=1)
```

```python
data_of_database_all_pages
```

```python
data_of_database_all_pages.info()
```

# Distribution of the dataset according to columns

## Distribution of the Experience of the Author's dataset


```python
data_of_database_all_pages["Experience Num Type"].value_counts().plot(kind='pie',figsize=(10, 15),autopct='%1.1f%%',textprops={'fontsize':15})
plt.legend()
plt.title("Experience type distribution")
```

## Conclusion: 
Most of our authors are veterans with years of experience

## Distribution of the datasets according to their last update in the dataset

```python
data_of_database_all_pages["Date Type"].value_counts().plot(kind='pie',figsize=(10, 15),autopct='%1.1f%%',textprops={'fontsize':15})
plt.legend()
plt.title("Date type distribution")
```

## Conclusion: 
most of our datasets have not been updated for at least a year

## Distribution of the datasets according to their size

```python
data_of_database_all_pages["File Size Type"].value_counts().plot(kind='pie',figsize=(10, 15),autopct='%1.1f%%',textprops={'fontsize':15})
plt.legend()
plt.title("File type distribution")
```

## Conclusion: 
Most of our datasets are in MB, which means that most of them are not big datasets

# Author's experience affects Usability and dataset size

```python
df = data_of_database_all_pages[data_of_database_all_pages["Experience Num"]>0].copy()
the_plot = df.groupby(["Experience Num Type"]).mean()[["Usability"]]
the_plot.plot()
```

```python
df = data_of_database_all_pages[data_of_database_all_pages["Experience Num"]>0].copy()
the_plot = df.groupby(["Experience Num Type"]).mean()[["size in Mb"]]
the_plot.plot()
```

## Conclusion: 
In regard of the information shown in the graphs, it is highly noticable that the less accomplished authors focus more on usability of the data while the more accomplished authors focus on the quantity of the information regarding the topic in the database.


# Columns affects Rating

# Searching for correlation between 2 columns

```python
corr_spearman = data_of_database_all_pages.corr(method='spearman')
fig = plt.figure(figsize = (20,10))
sns.heatmap(corr_spearman, annot=True, cmap='RdYlGn', vmin=-1, vmax=+1)
plt.title('Spearman Correlation of details of the databases ')
plt.show()
```

## Conclusions: 
As we can see there are 2 squares that have good correlation:
1. the left square represent the good correlation of the dataset's attribute
2. the right square represent the good correlation of the author's attribute
##### so we focused on them

# The correlation of Dataset: 

```python
features_databases = ['Rating', 'Views', 'Downloads',  'Notebooks', 'Topics']
```

```python

corr_spearman = data_of_database_all_pages[features_databases].corr(method='spearman')

fig = plt.figure(figsize = (8,6))
sns.heatmap(corr_spearman, annot=True, cmap='RdYlGn', vmin=-1, vmax=+1)
plt.title('Spearman Correlation of details of the databases ')
plt.show()
```

```python
df = data_of_database_all_pages[data_of_database_all_pages[features_databases]>0].copy()
sns.pairplot(df[features_databases])
plt.grid()
```

```python

df = data_of_database_all_pages.copy()
df.loc[(df.Rating <= 0) | (df.Rating > 9000 ),"Rating"]= np.nan
df.dropna(inplace=True)

plotR_V_D_N_T=df.groupby(['Rating']).mean()[['Views','Downloads','Notebooks','Topics']]
plotR_V_D_N_T.plot() 
```

```python
plotR_D_N_T=df.groupby(['Rating']).mean()[['Downloads','Notebooks','Topics']]
plotR_D_N_T.plot()
```

```python
plotR_N_T=df.groupby(['Rating']).mean()[['Notebooks','Topics']]
plotR_N_T.plot()
```

```python

plotR_T=df.groupby(['Rating']).mean()[['Topics']]
plotR_T.plot()
```

## conclusions:
* As we expected there is a noticable ascension as a dependency on Experience of the author.
* the reason for the "noises" in the plots can be explained as a result of outliers in the data.
